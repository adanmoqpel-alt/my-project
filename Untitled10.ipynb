```python


import sympy as sp
import numpy as np

def read_matrix():
 print("Enter number of rows and columns==> ")
 r,c=map(int,input().split())

 print("Enter matrix row by row")
 values=[]
 while len(values) < r*c:

    values+=[sp.Rational(v) for v in input().split()]
 #arr=np.array(values[:r*c]).reshape((r,c))
 arr=sp.Matrix(r,c,values)
 return arr

def menu():
 print( "\n------Choose from the menu------>>")
 print("1--> Row Echelon Form (REF)")
 print("2--> Reduced Row Echelon Form (RREF)")
 print("3--> Matrix addition ")
 print("4--> Matrix subtraction")
 print("5--> Scalar multiplication")
 print("6--> Matrix multiplication")
 print("7--> Determinant")
 print("8--> Creamer's rule")
 print("9--> Linear independence check")
 print("10---> Basis and dimension")
 print("11---> Egenvalues and Egenvectors")
 print("12---> Diagonalization")
 print("0---> EXIT")

def main():
 A=read_matrix()
 while True:
  menu()
  choice=int(input("choose an option: "))

  if choice==1:
    print("REF:")
    print(A.echelon_form())

  elif choice==2:

    print("RREF")
    print(A.rref()[0])

  elif choice==3:
    print("Enter second matrix B")
    B=read_matrix()
    if A.rows==B.rows and A.cols==B.cols:
     print("A+B=")
     print(A+B)
    else:
     print("Error!! matrices must have the same dimention ")

  elif choice==4:
    print("Enter second matrix B" )
    B=read_matrix()
    if A.rows==B.rows and A.cols==B.cols:
      print("A-B=")
      print(A-B)
    else:
     print("Error the matrices must have the same dimention")

  elif choice==5:
   S=float(input("Enter Scalar S :"))
   print("S*A=")
   print(S*A)

  elif choice==6:
   print("Enter second matrix B:")
   B=read_matrix()
   if A.cols==B.rows:
    print("A*B=")
    print(A*B)
   else:
    print("A'S columns must equals B's rows!!")

  elif choice==7:
   if A.rows==A.cols:
    print("det(A)=",A.det())
   else:
    print(" the matrix must be square!!")

  elif choice==8:
   print("Enter vector b")
   b=[]
   for i in range(A.rows):
    b.append(float(input(f"b[{i}]=")))
   b=sp.Matrix(b)
   print("Souliotion by creamer's rule")
   print(A.solve(b))


  elif choice==9:
   print("Cheaking linear inepndent :")
   print("Rank:",A.rank())
   if A.rank()==A.cols:
    print("columns are linearly independent")
   else:
    print("columns are linearly dependent")

  elif choice==10:
   print("\n Row space basis:",A.rowspace())
   print("\n Column space basis:",A.columnspace())
   print("\n Dimension:",A.rank())

  elif choice==11:
   print("Eginvalus:",A.eigenvals())
   print("Eginvectors:",A.eigenvects())

  elif choice==12:
   print("Diagonalizition:")
   try:
      P,D=A.diagonalize()
      print("P=")
      print(P)
      print("D=")
      print(D)
   except:
     print("Matrex is not diagonalizable")

  elif choice==0:
   break
  else:
   print('Invalid choice')

if __name__=="__main__":
  main()
```
