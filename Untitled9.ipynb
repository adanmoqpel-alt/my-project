```python
import sympy as sp

def solve_first_order():
    print("\n--- First ):Order ODE ---")
    print("Choose type:")
    print("1) Linear  y' + p(x)*y = q(x)")
    print("2) Separable  y' = f(x)*g(y)")
    t = input("Enter choice: ")

    x = sp.symbols('x')
    y = sp.Function('y')

    if t == "1":
        print("Enter p(x): ")
        p = sp.sympify(input())
        print("Enter q(x): ")
        q = sp.sympify(input())

        ode = sp.diff(y(x), x) + p*y(x) - q
        sol = sp.dsolve(ode)
        print("\nSolution:")
        print(sol)

    elif t == "2":
        print("Enter f(x): ")
        f = sp.sympify(input())
        print("Enter g(y): ")
        gy = sp.sympify(input())

        ode = sp.diff(y(x), x) - f*gy
        sol = sp.dsolve(ode)
        print("\nSolution:")
        print(sol)

    else:
        print("Invalid choice")


def solve_second_order():
    print("\n--- Second Order ODE ---")
    print("Equation form: a*y'' + b*y' + c*y = r(x)")

    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))

    print("Choose r(x):")
    print("1) 0")
    print("2) constant")
    print("3) exp(kx)")
    print("4) sin(kx)")
    print("5) cos(kx)")
    t = input("Enter choice: ")

    x = sp.symbols('x')
    y = sp.Function('y')

    if t == "1":
        r = 0
    elif t == "2":
        r = sp.sympify(input("Enter constant: "))
    elif t == "3":
        k = sp.sympify(input("Enter k: "))
        r = sp.exp(k*x)
    elif t == "4":
        k = sp.sympify(input("Enter k: "))
        r = sp.sin(k*x)
    elif t == "5":
        k = sp.sympify(input("Enter k: "))
        r = sp.cos(k*x)
    else:
        print("Invalid choice")
        return

    ode = a*sp.diff(y(x), x, 2) + b*sp.diff(y(x), x) + c*y(x) - r
    sol = sp.dsolve(ode)
    print("\nSolution:")
    print(sol)


def main():
    print("Differential Equation Solver")
    print("1) First order")
    print("2) Second order")
    choice = input("Enter type: ")

    if choice == "1":
        solve_first_order()
    elif choice == "2":
        solve_second_order()
    else:
        print("Invalid choice")


if __name__ == "__main__":
    main()














```
